# Análise de gastos Públicos de Deputados

## Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import locale
locale.setlocale(locale.LC_ALL, '')
```

## Definindo informações da tabela

```python
gastos= pd.read_csv(r'C:\Users\Renato\tabelaComparativa.csv', error_bad_lines=False, low_memory=False, delimiter=';', encoding='latin-1', decimal=',')
```

## Colunas

```python
gastos.columns
```

### Gastos

```python
gastos['Gastos'].value_counts()
```

### Nomes

```python
gastos.groupby("Nome").mean()["Gastos"].sort_values().nlargest(25)
```

# Gráficos

## Gráfico de gastos individuais

```python
x=gastos.groupby("Nome").mean()["Gastos"].sort_values().nlargest(25)
```

```python
x.plot(kind='bar' , subplots=True, title="Deputados que mais gastaram em 2019 (de Janeiro a 15 de Junho)", figsize=(7,6))
```

## Gráfico de gastos por partido

```python
y=gastos.groupby("Partido").mean()["Gastos"].sort_values().nlargest(27)
```

```python
y.plot(kind='barh' , subplots=True, title="Partidos que mais gastaram dinehiro público em 2019 (de Janeiro a 15 de Junho)", figsize=(7,6))
```

## Gráfico de distribuição de gastos por UF

```python
z=gastos.groupby("UF").mean()["Gastos"].sort_values().nlargest(28)
```

```python
z.plot(kind='pie' , subplots=True, title="Distribuição dos gastos por estado (de Janeiro a 15 de Junho)", figsize=(7,6))
```

## Gráfico que dos Estados que mais gastaram dinheiro público

```python
z.plot(kind='bar' , subplots=True, title="Estado que mais gastaram dinehiro público em 2019 (de Janeiro a 15 de Junho)", figsize=(7,6))
```
